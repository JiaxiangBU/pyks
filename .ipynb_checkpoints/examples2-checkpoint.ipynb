```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('refs/two_class_example.csv')
```

```python
fpr, tpr, thres = roc_curve(data.y, data.yhat)
ks = tpr - fpr
ks_max = np.max(ks)
print(ks_max)
```

```python
plt.plot(thres, ks)
plt.plot(thres, tpr)
plt.plot(thres, fpr)
plt.xlabel('Cutoff')
plt.ylabel('KS')
plt.title(str(ks_max))
plt.xlim(0,1)
plt.show()
plt.clf()
```
